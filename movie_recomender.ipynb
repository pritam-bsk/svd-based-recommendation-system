```python
import ast
def convert(obj):
    lis= []
    for i in ast.literal_eval(obj):
        lis.append(i['name'])
    return lis
```

```python
mvi['genres'] = mvi.genres.apply(convert)
mvi.keywords = mvi['keywords'].apply(convert)
```

```python
def convert2(obj):
    lis= []
    temp= 0
    for i in ast.literal_eval(obj):
        if temp == 3:
            break
        lis.append(i['name'])
        temp += 1
    return lis
```

```python
mvi.cast = mvi['cast'].apply(convert2)
```

```python
mvi.head()
```

```python
def convert3(obj):
    lis= []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            lis.append(i['name'])
    return lis
```

```python
mvi.head(1)
```

```python
mvi['crew'] = mvi['crew'].apply(convert3)
```

```python
def coll(obj):
    lis= []
    for i in obj:
        lis.append(i.replace(" ",""))
    return lis
```

```python
mvi['keywords']= mvi['keywords'].apply(coll)
mvi['cast']= mvi['cast'].apply(coll)
mvi['crew']= mvi['crew'].apply(coll)
```

```python
mvi['genres']= mvi['genres'].apply(coll)
```

```python
mvi.head(1)
```

```python
mvi.overview= mvi.overview.apply(lambda x: x.split())
```

```python
#mvi['tag'] = mvi['overview']+mvi['title']+mvi['overview']+mvi[]
mvi['tags'] = mvi['overview'] + mvi['genres'] + mvi['keywords'] + mvi['cast'] + mvi['crew']
```

```python
mvi.head(1)
```

```python
new_df = mvi.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_df.head()
```

```python
new_df.shape
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df.iloc[1].tags
```

```python
from nltk.stem.porter import PorterStemmer
ps= PorterStemmer()
```

```python
lis= []
for g,i in enumerate(new_df['tags']):
    tag = i.split()
    for j,k in enumerate(tag):
        tag[j]= ps.stem(k)
    string = " ".join(tag)
    lis.append(string)
```

```python
new_df['tags'] = lis
```

```python
new_df['tags'][1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity= cosine_similarity(vectors)
```

```python
similarity
```

```python
sorted(list(enumerate(similarity[56])),reverse=True)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```

```python
recommend('Falcon Rising')
```
